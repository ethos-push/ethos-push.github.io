```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('./general_5000_32_32_synth.csv')
df['contactSignificantContext'] = df['contactSignificantContext'].replace({'false':'False', 'true':'True'})
cols_to_round = ['numberUpdates', 'timeAppLastUsed', 'firstToLast', 'date_hour', 'date_min']
for col in cols_to_round:
    df[col] = df[col].astype('int64')
for col in df.columns:
    df[col] = df[col].astype('str')
df.date_dayofweek = df.date_dayofweek.replace({'0':'Mon', '1':'Tues', '2':'Wed', '3':'Thur', '4':'Fri', '5':'Sat', '6':'Sun'})
    
```
